```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) 
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) 

x2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12]) 
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(x, y, label='Day One', color='blue')

plt.scatter(x2, y2, label='Day Two', color='red')

plt.xlabel('Age of Cars')
plt.ylabel('Car Speed')
plt.title('Scatter Plot - Day One vs Day Two')
plt.legend()

plt.savefig('Day1_and_Day2.png', format='jpg') 
plt.show()
```
