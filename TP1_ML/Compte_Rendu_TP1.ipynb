```python
# Lire les données à partir du csv 'salesData.csv'
df = pd.read_csv('salesData.csv')

# Extraire les colonnes 'month_number' et 'total_profit'
month_numbers = df['month_number']
total_profit = df['total_profit']

# Créer un graphique pour la visualisation selon l'énoncé
plt.plot(month_numbers, total_profit, linestyle='--', color='red', marker='o',linewidth=3)

# Étiquettes des axes et légende
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(['Company Sales data of last year'], loc=4)

#Sauvegarde et affichage
plt.savefig('Question2.png', format='jpg') 
plt.show()
```

<div style="font-family: Open Sans; color: red; font-size:17px;; ">
    <b>3- Affichez toutes les données de ventes de produits (sale data) en utilisant un plot 
multiligne.</b>
</div>

```python
# Lire les données à partir du csv 'salesData.csv'
df = pd.read_csv('salesData.csv')

# Exclure les deux dernières colonnes
df = df.iloc[:, :-2]

#Boucler sur les colonnes et créer un graphique pour la visualisation
for column in df.columns[1:]:
    plt.plot(df['month_number'], df[column], marker='o', label=column)

# Étiquettes des axes et légende   
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.title('Ventes de Produits par Mois')
plt.legend(loc=2)

#Sauvegarde et affichage
plt.savefig('Question3.png', format='jpg') 
plt.show()

```

<div style="font-family: Open Sans; color: red; font-size:17px;; ">
    <b>4- Affichez les données de ventes de dentifrice (toothpaste sales data) de chaque mois à 
l'aide d'un nuage de points (scatter plot).</b>
</div>

```python
# Lire les données à partir du csv 'salesData.csv'
df = pd.read_csv('salesData.csv')

# Extraire la colonne 'ventes de dentifrice'
toothpaste_sales = df['toothpaste']

# Créer un nuage de points
plt.scatter(df['month_number'], toothpaste_sales, color='blue', marker='o', label='Tooth paste Sales Data')

# Étiquettes des axes, légende et titre du graphique
plt.xlabel('Month number')
plt.ylabel('Number of Units Sold')
plt.legend()
plt.title('Tooth paste Sales Data')

plt.grid(True)  

# Sauvegarde et Affichage
plt.savefig('Question4.png', format='jpg') 
plt.show()
```

<div style="font-family: Open Sans; color: red; font-size:17px;; ">
    <b>5- Calculez les données de vente totales de l'année dernière pour chaque produit et 
affichez-les à l'aide d'un graphique à secteurs (pit chart). Le Nombre d'unités vendues 
par an pour chaque produit est affiché en pourcentage.</b>
</div>

```python
# Lire les données à partir du csv 'salesData.csv'
df = pd.read_csv('salesData.csv')

#Calculer les données de vente totales de l'année dernière pour chaque produit
product_columns = df.columns[1:-2] 
total_sales_per_product = df[product_columns].sum()

# Créer un graphique pour la visualisation selon l'énoncé
plt.pie(total_sales_per_product, labels=total_sales_per_product.index, autopct='%1.1f%%')
plt.legend(loc=4)

# Titre du graphique
plt.title('Sales Data')

#Sauvegarde et affichage
plt.savefig('Question5.png', format='jpg')
plt.show()
```
