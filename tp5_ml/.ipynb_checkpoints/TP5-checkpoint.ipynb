```python
# Importing required packages
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
#2. Génération des données/ étiquettes synthétiques 
# Generating synthetic data and labels
features, true_labels = make_blobs(
    n_samples=200,
    centers=3,
    cluster_std=2.75,
    random_state=42
)
```

```python
#4. Prétraitement des données 
# Standardizing the features
# pour que les données aient une moyenne de 0 et un écart type de 1. 

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

```

```python
# 5. Génération des clusters 

# Creating a KMeans instance with specified parameters
kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42
)

# Fitting the KMeans model to the scaled features
kmeans.fit(scaled_features)

# a) Displaying the sum of squared distances (inertia)
print("a) Inertia (Sum of Squared Distances to Closest Cluster Center):", kmeans.inertia_)

# b) Displaying the coordinates of the final cluster centers
print("b) Final Cluster Centers:\n", kmeans.cluster_centers_)

# c) Displaying the number of iterations run by the K-Means algorithm
print("c) Number of Iterations:", kmeans.n_iter_)

# d) Displaying cluster labels for the first 5 data points
print("d) Cluster Labels for the first 5 data points:", kmeans.labels_[:5])
```

```python
# Define KMeans parameters
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}
sse = [] #somme des carrés intra-cluster
#Pour chaque cluster, calcule la distance au carré entre chaque point du cluster et le centre de ce cluster.
#Additionne ces distances au carré pour tous les clusters.

# Iterate over a range of k values (1 to 10)
for k in range(1, 11):
    # Create a KMeans instance with the current k
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)   
    # Fit the KMeans model to the scaled features
    kmeans.fit(scaled_features)   
    # Append the SSE to the list
    sse.append(kmeans.inertia_)
    
# a) Plotting the SSE in function of the number of clusters (Elbow Method)
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.show()

# Utiliser KneeLocator pour trouver le coude (det k optimale)
kl = KneeLocator(range(1, 11), sse, curve="convex", direction="decreasing")

# Afficher la valeur optimale de k (coude)
optimal_k = kl.elbow
print("La meilleure valeur de k (coude) :", optimal_k)
```

```python
#La meilleure valeur de k est généralement identifiée au niveau du "coude" de la courbe
#là où la diminution de l'erreur devient moins prononcée.
```

```python
from sklearn.metrics import silhouette_score

# Initialisation de la liste pour stocker les coefficients de silhouette
silhouette_coefficients = []

# Boucle sur les valeurs de k
for k in range(2, 11):
    # Création d'une instance KMeans avec la valeur actuelle de k
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    
    # Ajuster le modèle KMeans aux features normalisées
    kmeans.fit(scaled_features)
    
    # Calcul du coefficient de silhouette
    # il mesure la similarité intra-cluster par rapport à la similarité inter-cluster.
    score = silhouette_score(scaled_features, kmeans.labels_)
    
    # Ajout du coefficient de silhouette à la liste
    silhouette_coefficients.append(score)

# a) Tracer le coefficient de silhouette en fonction du nombre de clusters
plt.figure(figsize=(8, 4))
plt.plot(range(2, 11), silhouette_coefficients, marker='o')
plt.title('Coefficient de Silhouette en fonction du nombre de clusters')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Coefficient de Silhouette')
plt.xticks(range(2, 11), fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
Méthode du Coude (Elbow Method) :

Objectif : Identifier le nombre optimal de clusters (k).
Mesure : Basée sur la variation de l'inertie ou de la somme des carrés intra-cluster (SSE).
Fonctionnement : On ajuste le modèle pour différents k et observe comment l'inertie change. On recherche le point sur le graphique où l'ajout d'un cluster ne réduit plus significativement l'inertie (coude).


Objectif : Mesurer à quel point chaque point de données est bien assorti à son cluster par rapport aux autres clusters.
Mesure : Basée sur la distance moyenne intra-cluster (a) et inter-cluster (b).
Fonctionnement : On calcule le coefficient de silhouette pour chaque point de données en utilisant la formule (b - a) / max(a, b).
Interprétation : Proche de 1 indique une bonne correspondance à son propre cluster et une mauvaise correspondance aux autres. Proche de 0 indique une situation d'incertitude, et proche de -1 indique une mauvaise correspondance.
Comparaison :

La méthode du coude aide à déterminer k en se basant sur l'inertie du modèle.
Le coefficient de silhouette fournit une mesure globale de la qualité des clusters, indépendamment du nombre de clusters.
Il est recommandé d'utiliser plusieurs critères, tels que le coude et le coefficient de silhouette, pour prendre une décision éclairée sur le nombre de clusters à utiliser dans une analyse de clustering.
```

```python
# Importer les packages
import matplotlib.pyplot as plt
from sklearn import datasets
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd
```

```python
# Charger les données iris
iris = datasets.load_iris()
X = iris.data
X = pd.DataFrame(X)

# Calcul de la matrice de distance
Z = linkage(X, method="ward")

# Afficher le dendrogramme
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
D = dendrogram(Z, truncate_mode="lastp", p=10, ax=ax)  # Utilisation de truncate_mode
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Number of points in node")
plt.ylabel("Distance")
plt.show()
```
